```python
programming_dictionary= {
    "Bug" : "Bug is an error occured in a program",
    "Function" : "A piece of programm that can be used over and over again",
    "Loop" : "Loop is a action of doing someting over and over again"
}

print(programming_dictionary["Loop"])

#empty dictionary
empty_dictionary ={}
empty_dictionary["Bug"]= "Bug is an error occured in a program"
print(empty_dictionary)

#loop through dictionary
for key in programming_dictionary:
  print(key)
  print(programming_dictionary[key])
```

```python
student_scores = {
    'Harry': 88,
    'Ron': 78,
    'Hermione': 95,
    'Draco': 75,
    'Neville': 60
}

# Create an empty dictionary to collect the new values.
student_grades = {}

# Loop through each key in the student_scores dictionary
for student in student_scores:

    #Get the value (student score) by using the key each time.
    score = student_scores[student]

    #Check what grade the score would get, then add it to student_grades
    if score >= 91:
        student_grades[student] = 'Outstanding'
    elif score >= 81:
        student_grades[student] = 'Exceeds Expectations'
    elif score >= 71:
        student_grades[student] = 'Acceptable'
    else:
        student_grades[student] = 'Fail'
    #print(score)
print(student_grades)
```
