```python
#caesar cipher
import string
alphabet = []
for i in range(ord('a'), ord('z') + 1):
    alphabet.append(chr(i))
#print(alphabet)

direction = input("Type 'decode' for decrypt and 'encode' for encrypt:\n").lower()
text = input("type your msg:\n").lower()
shift = int(input("how many shifts you want ?\n"))

def encrypt(original_text, shift_amt):
  cipher_text = "" #hold the letters after shifting
  for letter in original_text:
    shifted_position = alphabet.index(letter) + shift_amt
    shifted_position %= len(alphabet)
    cipher_text += alphabet[shifted_position]
  print(f"here is the encoded result:{cipher_text}\n")
encrypt(original_text = text , shift_amt =  shift)
```
