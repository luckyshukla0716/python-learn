**Project: Tip calculator**

```python
print("Welcome to the tip calculator")
no_of_people = input("How many people to split the bill?: ")
bill = float(input("What was the total bill?:"))
tip =int(input("how much you like to give tip? 10,15,20?:"))
tip_per= tip/100
total_amt_topay= bill+tip_per
print(f"total amount to pay {total_amt_topay}")
total_amt_byeach= total_amt_topay/int(no_of_people)
print(f"each person has to pay {total_amt_byeach}")
print(round(total_amt_byeach))
print(round(total_amt_byeach,2))

```
