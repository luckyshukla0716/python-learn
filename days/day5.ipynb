```python
queue = []

# Enqueue
queue.append('A')
queue.append('B')
queue.append('C')
print("Queue: ", queue)

# Peek
frontElement = queue[0]
print("Peek: ", frontElement)

# Dequeue
poppedElement = queue.pop(0)
print("Dequeue: ", poppedElement)

print("Queue after Dequeue: ", queue)

# isEmpty
isEmpty = not bool(queue)
print("isEmpty: ", isEmpty)

# Size
print("Size: ", len(queue))
```

```python
import random

print("Welcome to password generator!")

# Get user input for password composition
alpha_count = int(input("How many alphabets do you want in your password?\n"))
num_count = int(input("How many numbers do you want in your password?\n"))
special_char_count = int(input("How many special characters do you want in your password?\n"))

# Initialize lists to hold all possible characters
all_alphabets = []
# Lowercase letters (ASCII 97-122)
for i in range(97, 123):
  all_alphabets.append(chr(i))
# Optional: Add uppercase letters (ASCII 65-90)
# for i in range(65, 91):
#   all_alphabets.append(chr(i))

all_numbers = []
# Digits 0-9
for i in range(0, 10):
  all_numbers.append(str(i))

all_special_chars = []
# Common special characters (ASCII 33-43: !, ", #, $, %, &, ', (, ), *, +)
for i in range(33, 44):
  all_special_chars.append(chr(i))

# Build the password character list based on user counts
password_characters = []

for _ in range(alpha_count):
    if all_alphabets: # Ensure list is not empty before choosing
        password_characters.append(random.choice(all_alphabets))

for _ in range(num_count):
    if all_numbers: # Ensure list is not empty before choosing
        password_characters.append(random.choice(all_numbers))

for _ in range(special_char_count):
    if all_special_chars: # Ensure list is not empty before choosing
        password_characters.append(random.choice(all_special_chars))

# Shuffle the generated characters to ensure randomness
random.shuffle(password_characters)

# Join the list to form the final password string
final_password = "".join(password_characters)

print(f"Your generated password is: {final_password}")
```
