```python
#nested loop--if elif
print("Welcome to rollercoster!")
height =int(input("enter your height :"))
age =int(input("enter your age :"))
bill=0
if height>=120: # Entry condition based on height
    print("Entry allowed")

    if age < 12:
        bill = 5
        print("entry fee will be $5")

    elif 12 <= age <= 18: # Age between 12 and 18 inclusive
        bill = 7
        print("entry fee will be $7")

    elif 45 <= age <= 55: # Age between 45 and 45 inclusive
            bill = 0
            print("entry fee will be $0")

    else: # Age greater than 18
        bill = 12
        print("entry fee will be $12")

    wants_photo = input("Do you want to have a photo take? Type y for yes and m for no ")
    if wants_photo == "y" (age != 45 and age != 55):
      #add $3 to the bill
      bill += 3
    print(f"your final bill is ${bill}")
else:
  print("Entry not allowed")

```

```python
#BMI Calculator with Interpretations
weight= float(input("enter your weight in kg :"))
height= float(input("enter your height in m :"))

bmi= weight/height**2
print(bmi)

if bmi < 18.5:
    print(f"Your BMI is {bmi:.2f}, you are underweight.")
elif 18.5 <= bmi < 25:
    print(f"Your BMI is {bmi:.2f}, you have a normal weight.")
elif bmi >= 25:
    print(f"Your BMI is {bmi:.2f}, you are slightly overweight.")
```

**Pizza Order** Program:

```python
print("Welcome to Pizza dieveries")
size = input("what size of pizza do you wants? S, M, and L?:").upper()
pepperoni = input("Do you want pepperoni? Y and N: ").upper()
extra_cheese = input("Do you want extra cheese? Y and N: ").upper()

bill = 0

if size == 'S':
  bill = 15
elif size == 'M':
  bill = 20
elif size == 'L':
  bill = 25

if pepperoni == 'Y':
  if size == 'S':
    bill += 2
  else:
    bill += 3

if extra_cheese == 'Y':
  bill += 1

print(f"Your final bill is ${bill}")
```
