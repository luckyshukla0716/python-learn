```python
import string

alphabet = []
for i in range(ord('a'), ord('z') + 1):
    alphabet.append(chr(i))
#print(alphabet)

logo = """
 ,adPPYba, ,adPPYYba,  ,adPPYba, ,adPPYba, ,adPPYYba, 8b,dPPYba,
a8"     "" ""     `Y8 a8P_____88 I8[    "" ""     `Y8 88P'   "Y8
8b         ,adPPPPP88 8PP"""""""  `"Y8ba,  ,adPPPPP88 88
"8a,   ,aa 88,    ,88 "8b,   ,aa aa    ]8I 88,    ,88 88
 `"Ybbd8"' `"8bbdP"Y8  `"Ybbd8"' `"YbbdP"' `"8bbdP"Y8 88
            88             88
           ""             88
                          88
 ,adPPYba, 88 8b,dPPYba,  88,dPPYba,   ,adPPYba, 8b,dPPYba,
a8"     "" 88 88P'    "8a 88P'    "8a a8P_____88 88P'   "Y8
8b         88 88       d8 88       88 8PP""""""" 88
"8a,   ,aa 88 88b,   ,a8" 88       88 "8b,   ,aa 88
 `"Ybbd8"' 88 88`YbbdP"'  88       88  `"Ybbd8"' 88
              88
              88
"""
print(logo)

def caesar(original_text, shift_amt, encode_or_decode):
  output_text =""
  if encode_or_decode == "decode":
    shift_amt *= -1

  for letter in original_text:

    if letter not in alphabet:
      output_text += letter
    else:
      shifted_position = alphabet.index(letter) + shift_amt
      shifted_position %= len(alphabet)
      output_text += alphabet[shifted_position]
  print(f"here is the {encode_or_decode}d result:{output_text}\n")

def get_restart_choice():
  choice = input("Type 'yes' if you want to go again otherwise type 'no':\n").lower()
  return choice == 'yes'

should_continue = True

while should_continue :
  while True:
    direction = input("Type 'decode' for decrypt and 'encode' for encrypt:\n").lower()
    if direction in ['encode', 'decode']:
      break
    else:
      print("Invalid input. Please enter 'encode' or 'decode'.")

  text = input("type your msg:\n").lower()

  while True:
    try:
      shift = int(input("how many shifts you want ?\n"))
      break
    except ValueError:
      print("Invalid input. Please enter a whole number for the shift.")

  caesar(original_text=text, shift_amt=shift, encode_or_decode=direction)

  should_continue = get_restart_choice()
  if not should_continue:
    print("Goodbye!")
```
