```python
#picking up a random word and checking ans

import random
word_list = ["apple", "lucky", "chintu"]

chosen_word =random.choice(word_list)
print(chosen_word)

placeholder = ""
word_length = len(chosen_word)
for position in range(word_length):
  placeholder += "-"
print(placeholder)


guess = input("choose your word:").lower()
#print(guess)

display = ""

for letter in chosen_word:
  #print(letter)
  if letter == guess :
    display += letter
  else:
    display += "-"

print(display)
```
