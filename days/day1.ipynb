simple interest= (p*t*r)/100   

P is the Principal amount
T is the Time period (in years)
R is the Rate of interest per annum

```python
def fun(p,t,r):
  return (p*t*r)/100

p,t,r= 3,25,8
res= fun(p,t,r)
print(res)
```

simple interest using Lambda function

```python
si = lambda p,t,r : (p*t*r)/100

p,t,r = 10,5,6

res= si(p,t,r)
print(res)
```

compound intrest:
A = P(1 + R/100) t
Compound Interest = A - P

```python
p=1200
r=5
t=12
amount =p*(1+r/100)**t
cp = amount-p
print(cp)
```
