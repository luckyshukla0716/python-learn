```python
def life_in_weeks(age):
    years_left = 90 - age
    weeks_left = years_left * 52
    print(f"You have {weeks_left} weeks left.")



life_in_weeks(56)
```

```python
def greet_with(name, location):
  print(f"Hello {name}!") #f-string
  print(f"What is it like in {location}?")
greet_with("Lucky", "USA")  #positional argument

def greet_with2(name, location, age):
  print(f"Hello {name}!")
  print(f"What is it like in {location}?")
  print(f"What is your {age}?")

greet_with2(name="Lucky", location="USA", age="29")  #keyword arguments
```

```python
def calculate_love_score(name1, name2):
    # Combine both names and convert to lowercase for case-insensitive counting
    combined_names = (name1 + name2).lower()

    # Count occurrences of letters in "TRUE"
    t = combined_names.count("t")
    r = combined_names.count("r")
    u = combined_names.count("u")
    e = combined_names.count("e")
    first_digit = t + r + u + e

    # Count occurrences of letters in "LOVE"
    l = combined_names.count("l")
    o = combined_names.count("o")
    v = combined_names.count("v")
    e = combined_names.count("e")
    second_digit = l + o + v + e

    # Combine the digits as strings, then convert to integer (or directly print the string format)
    love_score = f"{first_digit}{second_digit}"
    print(love_score)


# Example call:
calculate_love_score("Kanye West", "Kim Kardashian")
```
